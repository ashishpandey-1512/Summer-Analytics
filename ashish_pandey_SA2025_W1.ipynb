### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = cars_df['mpg'].median()

above_median_cars = cars_df[cars_df['mpg'] > median_mpg]

max_ratio = above_median_cars['hp_to_weight'].max()

best_cars = above_median_cars[above_median_cars['hp_to_weight'] == max_ratio]
best_cars
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import seaborn as sns

avg_mpg = cars_df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.xticks(avg_mpg['model_year'].unique())  # Show all years on x-axis
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index to get 'name' as a regular column if needed
df = cars_df.reset_index()

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    hue_order=['japan', 'europe', 'usa'],
    palette='Set2',
    alpha=0.7,
    edgecolor='k'
)

plt.title('Car Weight vs Horsepower, Colored by Origin & Sized by MPG', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
